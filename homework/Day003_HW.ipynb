```python
import xml.dom.minidom
doc = xml.dom.minidom.parse("/Users/jordan/2nd-PyCrawlerMarathon/2nd-PyCrawlerMarathon/data/sample.xml")
print(doc.getElementsByTagName("Title")[0].firstChild.nodeValue)
chapters = doc.getElementsByTagName("Chapter")
for chapter in chapters:
    print (chapter.getAttribute('name'), chapter.firstChild.nodeValue)
```

```python
import xml.etree.ElementTree as ET
tree = ET.parse("/Users/jordan/2nd-PyCrawlerMarathon/2nd-PyCrawlerMarathon/data/sample.xml") 
root = tree.getroot()
print(root[0].text)
chapters = root[2]
for chapter in chapters:
    print (chapter.attrib['name'], chapter.text)
```
